```python
from langchain.agents import create_agent
from langchain_google_genai  import ChatGoogleGenerativeAI
from dotenv import load_dotenv
from langchain_mcp_adapters.client import MultiServerMCPClient
from langchain.messages import HumanMessage
```

```python
load_dotenv(override=True)
```

```python
mcp_client = MultiServerMCPClient({
    "mcp-server": {
        "transport": "streamable_http",
        "url": "http://localhost:24000/mcp"
    }
})
```

```python
tools = await mcp_client.get_tools()
```

```python
tools
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0
)
```

```python
agent = create_agent(
    model=llm,
    tools=tools,
    system_prompt="Answer user questions using the tools available."
)
```

```python
resp = await agent.ainvoke({
    "messages" : [
        HumanMessage("Qule est le salaire de Aymen")
    ]
})
```

```python
print((resp["messages"][-1].content[0].get("text")))
```

```python
resp = await agent.ainvoke({
    "messages" : [
        HumanMessage("Qule est la valeur de l'acation de Microsoft aujourd'hui")
    ]
})
```
